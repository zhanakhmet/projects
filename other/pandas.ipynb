```python
import pandas as pd

df = pd.read_excel('events.xlsx')

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

df = df.loc[df['К какому типу относится?'] != 'Дефект']
df # список без показателя дефект
kf = df.groupby(['Год', 'Месяц', 'День'])['К какому типу относится?'].count()
kf #final kf
```
